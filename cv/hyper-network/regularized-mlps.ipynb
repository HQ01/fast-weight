```python
import matplotlib.pylab as pl
%matplotlib inline
import cPickle as pickle
```

```python
accuracy_table, progress_table = {}, {}
accuracy_table['baseline'], progress_table['baseline'] = pickle.load(open('info/cifar-baseline-mlp-3-1536', 'rb'))
accuracy_table['dropout'], progress_table['dropout'] = pickle.load(open('info/dropping-out-mlp-3-3072', 'rb'))
accuracy_table['pooling'], progress_table['pooling'] = pickle.load(open('info/element-wise-stochastic-pooling-mlp-3-3072', 'rb'))
```

```python
for key, value in progress_table.items():
    accuracy = accuracy_table[key]
    training_accuracy = value['training_accuracy']
    pl.plot(range(len(training_accuracy)), training_accuracy, label='%s training' % key)
    validation_accuracy = value['validation_accuracy']
    pl.plot(range(len(validation_accuracy)), validation_accuracy, label='%s validation#%f' % (key, accuracy))
pl.legend(framealpha=0, loc='best')
pl.grid()
pl.xlabel('epoch')
pl.ylabel('validation_accuracy')
```
