```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
n_layers = 9
accuracy_table, progress_table = {}, {}
# for decaying_rate in (0, 0.5, 1.0, 1.5, 2.0, 2.5):
for decaying_rate in (0, 0.5, 1, 1.5):
    accuracy_table[decaying_rate], progress_table[decaying_rate] = \
        pickle.load(open('info/cifar-decayed-memory-network-%d-decaying-rate-%f' % (n_layers, decaying_rate)))
```

```python
for key, value in sorted(accuracy_table.items()):
    progress = progress_table[key]
    validation_accuracy = progress['validation_accuracy']
    label = '%d layers decaying_rate %f#%f' % (n_layers, key, value)
    pl.plot(range(len(validation_accuracy)), validation_accuracy, label=label)
pl.legend(framealpha=0, loc=0, fontsize='small')
pl.grid()
pl.xlabel('epoch')
pl.ylabel('validation_accuracy')
```
