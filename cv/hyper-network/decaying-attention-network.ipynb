```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
accuracy_table, progress_table = {}, {}
for N in (3,):
    accuracy_table[N], progress_table[N] = {}, {}
    for decaying_rate in (0.0, 0.5, 1.0):
#     for decaying_rate in (0.0, 0.5, 1.0, 1.5, 2.0, 2.5, 3.0):
        accuracy, progress = \
            pickle.load(open('info/cifar-decayed-memory-network-%d-decaying-rate-%6f' % (N, decaying_rate), 'rb'))
        accuracy_table[N][decaying_rate], progress_table[N][decaying_rate] = accuracy, progress
```

```python
for N in (3,):
    for decaying_rate in (0.0, 0.5, 1.0):
        accuracy, progress = accuracy_table[N][decaying_rate], progress_table[N][decaying_rate]
        training_accuracy = progress['training_accuracy']
        label = 'N=%d decaying_rate=%f#%f' % (N, decaying_rate, accuracy)
        pl.plot(range(len(training_accuracy)), training_accuracy, label=label)
pl.legend(framealpha=0, loc=0, fontsize='small')
pl.xlabel('epoch')
pl.ylabel('training_accuracy')
```

```python
for N in (3,):
    for decaying_rate in (0.0, 0.5, 1.0):
        accuracy, progress = accuracy_table[N][decaying_rate], progress_table[N][decaying_rate]
        validation_accuracy = progress['validation_accuracy']
        label = 'N=%d decaying_rate=%f#%f' % (N, decaying_rate, accuracy)
        pl.plot(range(len(validation_accuracy)), validation_accuracy, label=label)
pl.legend(framealpha=0, loc=0, fontsize='small')
pl.xlabel('epoch')
pl.ylabel('validation_accuracy')
```
