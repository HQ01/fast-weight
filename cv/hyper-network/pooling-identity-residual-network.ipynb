```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
for N in range(0, 11):
    normal_accuracy, normal_progress = pickle.load(open('info/triple-state-normal-residual-network-%d' % (N + 1), 'rb'))
    pooling_accuracy, pooling_progress = \
        pickle.load(open('info/pooling-identity-residual-network-N-%d' % (N), 'rb'))
    pl.figure(figsize=(4, 3))
    pl.plot(
        range(len(normal_progress['validation_accuracy'])),
        normal_progress['validation_accuracy'],
        label='normal#%f' % normal_accuracy
    )
    pl.plot(
        range(len(pooling_progress['validation_accuracy'])),
        pooling_progress['validation_accuracy'],
        label='pooling#%f' % pooling_accuracy
    )
    pl.grid()
    pl.title('%d residual modules per transition' % (N + 1), fontsize='medium')
    pl.xlabel('epoch')
    pl.ylabel('validation accuracy')
    pl.legend(framealpha=0, loc='best', fontsize='medium')
```
