```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
one_accuracy, one_progress = pickle.load(open('info/one-activation-network-3'))
one_accuracy
```

```python
dual_accuracy, dual_progress = pickle.load(open('info/dual-activation-network-3', 'rb'))
dual_accuracy
```

```python
pl.plot(range(len(one_progress['validation_accuracy'])), one_progress['validation_accuracy'], label='one')
pl.plot(range(len(dual_progress['validation_accuracy'])), dual_progress['validation_accuracy'], label='dual')
pl.grid()
pl.xlabel('epoch')
pl.ylabel('validation accuracy')
pl.legend(framealpha=0, loc='best')
```
