```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
accuracy_table = {}
progress_table = {}
for function in ('average', 'latest'):
    accuracy, progress = pickle.load(open('info/unattentioned-network-times-8-function-%s' % function, 'rb'))
    accuracy_table[function] = accuracy
    progress_table[function] = progress
```

```python
for key, value in progress_table.items():
    pl.plot(
        range(len(value['validation_accuracy'])),
        value['validation_accuracy'],
        label='%s %f' % (key, accuracy_table[key])
    )
    pl.ylim(0, 1)
    pl.legend(framealpha=0, loc='best')
    pl.grid()
```
