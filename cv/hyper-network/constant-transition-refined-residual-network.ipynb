```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
accuracy_table, progress_table = [], []
for N in range(21):
    accuracy, progress = \
        pickle.load(open('info/triple-state-refined-residual-network-%d-constant-transition' % (N), 'rb'))
    accuracy_table.append(accuracy)
    progress_table.append(progress)
```

```python
pl.plot(range(len(accuracy_table)), accuracy_table, linewidth=1)
pl.title('test accuracy of refined residual networks')
pl.xlabel('refinement modules per transition')
pl.ylabel('test accuracy')
pl.grid()
```

```python
def plot_progress(progress_table, entry):
    pl.figure(figsize=(9, 9))
    for index, progress in enumerate(progress_table):
        table = progress[entry]
        pl.plot(range(len(table)), table, label=str(index + 1))
    pl.title(entry.replace('_', ' '))
    pl.xlabel('epoch')
    pl.ylabel('value')
    pl.legend(framealpha=0, loc='best')
```

```python
plot_progress(progress_table, 'training_loss')
```

```python
plot_progress(progress_table, 'training_accuracy')
```

```python
plot_progress(progress_table, 'validation_accuracy')
```

```python
plot_progress(progress_table, 'validation_loss')
```
