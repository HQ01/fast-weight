```python
import cPickle as pickle
import matplotlib.pylab as pl
%matplotlib inline
```

```python
N_LAYERS = 10
info = {}
info['shared weight baseline'] = pickle.load(open('info/rescaled-mnist-baseline-network-%d-weight-sharing' % N_LAYERS, 'rb'))
info['private weight baseline'] = pickle.load(open('info/rescaled-mnist-baseline-network-%d-private-weight' % N_LAYERS, 'rb'))
```

```python
for key, value in info.items():
    accuracy, progress = value
    pl.plot(range(len(progress['validation_accuracy'])), progress['validation_accuracy'], label='%s#%f' % (key, accuracy))
pl.legend(framealpha=0, loc='best')
pl.xlabel('epoch')
pl.ylabel('validation accuracy')
```
