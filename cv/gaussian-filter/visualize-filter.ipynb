```python
from joblib import load
params = load('parameters/msp-cnn-2-filters-16-units-multi-scale-mnist')
```

```python
import matplotlib.pylab as pl
%matplotlib inline
```

```python
for index in range(3):
    param_name = 'convolution%d_weight' % index
    param = params[param_name].asnumpy()
    OUT, IN, _, _ = param.shape
    for i in range(OUT):
        for j in range(IN):
            pl.figure()
            pl.title('convolution %d filter %d %d' % (index, i, j))
            pl.imshow(param[i][j], cmap='gray')
```
