```python
from joblib import load
identifier = 'info/msp-cnn-%d-filters-64-units-multi-scale-mnist'
msp_table = tuple(load(identifier % (n_filters * 2))[1] for n_filters in range(1, 9))
identifier = 'info/referential-cnn-%d-filters-multi-scale-mnist'
referential_table = tuple(load(identifier % (n_filters * 2))[1] for n_filters in range(1, 9))
```

```python
import matplotlib.pylab as pl
%matplotlib inline
pl.plot(range(2, 18, 2), msp_table, label='multi-scale pooling cnn')
pl.plot(range(2, 18, 2), referential_table, label='conventional cnn')
pl.legend(framealpha=0, loc=0)
```

```python
for n_filters in range(2, 18, 2):
    pl.figure()
    msp = load('info/msp-cnn-%d-filters-64-units-multi-scale-mnist' % n_filters)[0]
    pl.plot(range(len(msp)), msp, label='multi-scale pooling cnn')
    reference = load('info/referential-cnn-%d-filters-multi-scale-mnist' % n_filters)[0]
    pl.plot(range(len(reference)), reference, label='conventional cnn')
    pl.legend(framealpha=0, loc=0)
    pl.title('%d filters' % n_filters)
```
