```python
from joblib import load
history_table = {'standard scale mnist' : [], 'multi-scale mnist' : []}
filter_table = range(1, 17)
for n_filters in filter_table:
    for key, value in history_table.items():
        identifier = 'conventional-cnn-%d-filters-%s' % (n_filters, key.replace(' ', '-'))
        value.append(load('info/%s' % identifier))
```

```python
import matplotlib.pylab as pl
%matplotlib inline
for key, value in history_table.items():
    pl.plot(filter_table, tuple(table[1] for table in value), 'o', label=key)
pl.legend(framealpha=0, loc=0)
pl.xlabel('number of filters')
pl.ylabel('test error')
```

```python
for n_filters, standard, multi in \
    zip(filter_table, history_table['standard scale mnist'], history_table['multi-scale mnist']):
    pl.figure()
    pl.plot(range(len(standard[0])), standard[0], label='standard scale mnist')
    pl.plot(range(len(multi[0])), multi[0], label='multi-scale mnist')
    pl.title(n_filters)
    pl.legend(framealpha=0, loc=0)
    pl.xlim(0, 25)
    pl.ylim(0, 1)
    pl.grid()
```
